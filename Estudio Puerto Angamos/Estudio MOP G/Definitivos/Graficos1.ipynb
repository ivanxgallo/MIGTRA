```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import time
import matplotlib 
import datetime
matplotlib.rc_file_defaults()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
params = {'xtick.labelsize': 10, 'ytick.labelsize': 10}
matplotlib.rcParams.update(params)
pd.set_option('display.max_columns', None)
```

## Primer Plot

```python
file_stats = "SQMStats_Pangamos_20210830-20220203v3.xlsx"
df = pd.read_excel(file_stats)
```

```python
df_mop = df[(df["Producto"].notna()) & (df["Producto"].str.contains("MOP")) & (df["Hora Ingreso a Puerto"].notna())]
df_mop.head(1)
```

```python
cols = df_mop.columns
cols = cols[~cols.str.contains("Unnamed")]
df_mop = df_mop[cols]
df_mop.head(1)
```

```python
#pd.set_option("display.max_columns", None)
pdf = []
rangos = [0, 15, 30, 45, 60, 75, 90, 120, 180, 240, 300, 360]
for i in range(1,len(rangos)):
    pdf.append(df_mop[(df_mop["Tiempo Total en Puerto"] > rangos[i-1]) & (df_mop["Tiempo Total en Puerto"] <= rangos[i])].shape[0])
    
pdf = np.array(pdf)
pdf
```

```python
tot = pdf.sum()
perc = 100*np.array(pdf)/tot
tot
```

```python
parameters = {'xtick.labelsize': 10, 'ytick.labelsize': 10}
plt.rcParams.update(parameters)
bins = np.array([r"$<$"+str(i) for i in rangos[1:]])
fig = plt.figure(figsize=(7,3), dpi = 250)

ax1 = fig.add_subplot(111)

ax1.set_title("Distribución del Tiempo de Descarga")
ax1.set_ylabel("N° de Viajes", fontsize = 13)
ax1.set_xlabel("Rango de Tiempo de Descarga (mins)", fontsize = 13)
ax1.bar(bins, pdf, label = "Viajes")
ax1.set_ylim([0,1000])


for i, label in enumerate(perc):
    plt.annotate(str(round(label,1))+"\%", (i-0.3, pdf[i]+7.9))

#plt.savefig("Graficos/Distribucion Tiempo Descarga.png", bbox_inches= "tight")   
```

## Segundo Plot


```python
# Reseteamos el indice del dataset
df_mop.reset_index(drop = True, inplace = True)
df_mop.tail(1)
```

```python
df_mop["Hora Ingreso a Puerto"] = pd.to_datetime(df_mop["Hora Ingreso a Puerto"])
```

```python
df_mop.head(1)
```

```python
t1 = datetime.time(0,0,0)
t2 = datetime.time(8,30,0)
t3 = datetime.time(10,0,0)
t4 = datetime.time(12,0,0)
t5 = datetime.time(14,0,0)
t6 = datetime.time(15,30,0)
t7 = datetime.time(17,0,0)
t8 = datetime.time(19,0,0)
t9 = datetime.time(21,0,0)
t10 = datetime.time(22,30,0)
t11 = datetime.time(23,59,59)


T = np.array([t2, t3, t4, t5, t6, t7, t8, t9, t10, t11])


turnos = np.array(["Pre", "T1A", "T1B", "T1C", "Almuerzo",
          "T2A", "T2B", "T2C", "T2D", "Post"])
```

```python
def Turno(t):
    for idx, val in enumerate(T):
        if t.time() < val:
            return turnos[idx]
```

```python
df_mop["Turno en Puerto"] = df_mop["Hora Ingreso a Puerto"].apply(lambda x: Turno(x))
```

```python
df_mop.head(3)
```

```python
N_truck = np.array([df_mop[df_mop["Turno en Puerto"] == i].shape[0] for i in turnos])
```

```python
T_prom = np.array([df_mop["Tiempo Total en Puerto"][df_mop["Turno en Puerto"] == i].mean() for i in turnos])
```

```python
Tprom = (N_truck*T_prom).sum()/N_truck.sum()
Tprom
```

```python
T_prom = np.array([df_mop["Tiempo Total en Puerto"][df_mop["Turno en Puerto"] == turnos[i]].mean() for i in range(len(turnos))])
T_prom
```

```python
parameters = {'xtick.labelsize': 10, 'ytick.labelsize': 10}
plt.rcParams.update(parameters)
fig = plt.figure(figsize=(7,3), dpi = 250)

ax1 = fig.add_subplot(111)

ax1.bar(turnos, N_truck, label = "N° Camiones", color = "deepskyblue")
ax1.set_ylim([0,700])
ax1.set_ylabel("N° Camiones", fontsize = 13)
ax1.set_xlabel("Turno", fontsize = 13)
ax1.legend(loc = "upper left")
ax1.set_xticks(turnos)
ax1.set_xticklabels(turnos, rotation = 0)


ax2 = ax1.twinx() 

ax2.plot(turnos, T_prom, "ro", label = "Tiempo Promedio \n Por Turno")
ax2.set_title("Tiempos de Descarga según Turno en Faena", fontsize = 15)
ax2.set_ylabel("Tiempo Promedio de Descarga (min)")
ax2.axhline(y = Tprom, c = "r", 
                linestyle = "--", linewidth = 1, 
            label = "Tiempo Promedio \n Total: "+ str(round(Tprom)) + " min.")
ax2.axhline(y = 60., c = "k", 
                linestyle = "--", linewidth = 1, label = "60 minutos")

ax2.set_ylim([50,135])
ax2.legend(fontsize = 8)



for i, label in enumerate(T_prom):
    plt.annotate(str(round(label,1)), (i-0.2, T_prom[i]+2.7))


#plt.savefig("Graficos/Tiempos de Descarga por Turno.png", bbox_inches= "tight")
```

## Tercer Plot

```python
df_mop.head(1)
```

```python
fechas = df_mop["Hora Ingreso a Puerto"].dt.date.unique()
```

```python
truck_day = np.array([df_mop[df_mop["Hora Ingreso a Puerto"].dt.date == i].shape[0] for i in fechas])
truck_day
```

```python
T_day = np.array([df_mop["Tiempo Total en Puerto"][df_mop["Hora Ingreso a Puerto"].dt.date == i].mean() for i in fechas])
T_day
```

```python
Tday = (truck_day*T_day).sum()/truck_day.sum()
Tday
```

```python
fechas_labels = [str(i) for i in fechas]
fechas_labels
```

```python
parameters = {'xtick.labelsize': 10, 'ytick.labelsize': 10}
plt.rcParams.update(parameters)
fig = plt.figure(figsize=(7,3), dpi = 250)

ax1 = fig.add_subplot(111)



ax1.bar(fechas_labels, truck_day, label = "N° Camiones", color = "deepskyblue")
ax1.set_ylabel("N° de Camiones", fontsize = 13)
ax1.set_xlabel("Fecha", fontsize  = 13)
ax1.set_xticks(fechas_labels)
ax1.set_xticklabels(fechas_labels, rotation = 40, fontsize = 6, fontdict = {"horizontalalignment": "right"}) #
ax1.legend(loc = "upper left")
ax1.set_ylim([0,200])

ax2 = ax1.twinx() 

ax2.plot(fechas_labels, T_day, "ro", label = "Tiempo Promedio \n Por Día")
ax2.set_title("Tiempos de Descarga según Día en Faena")
ax2.set_ylabel("Tiempo Promedio de Descarga (min)", fontsize = 11)
ax2.axhline(y = Tday, c = "r", 
                linestyle = "--", linewidth = 1, 
            label = "Tiempo Promedio \n Total: "+ str(round(Tday)) + " min.")
ax2.axhline(y = 60., c = "k", 
                linestyle = "--", linewidth = 1, label = "60 minutos")

ax2.set_ylim([0,300])
ax2.legend(fontsize = 7)


for i, label in enumerate(T_day):
    plt.annotate(str(round(label)), (i-0.45, T_day[i]+7.5), fontsize = 7)


#plt.savefig("Graficos/Tiempos de Descarga por Dia.png", bbox_inches= "tight")
```
