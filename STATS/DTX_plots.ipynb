```python
import os

def get_csv_file_names(directory):
    file_names = []
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)) and filename.endswith('.csv'):
            file_names.append(filename)
    return file_names
```

```python
directory = "stats_"
files = get_csv_file_names(directory)
files, len(files)
```

```python
clientes = [
    "coseducam2",
    "agrosuper_tp",
    "agrosuper4",
    "greenamerica2",
    "abastible_e",
    "abastible3",
    "sqmc3",
    "sqmpt",
    "sqmraevpp",
    "codelcodsal2",
    "codelcodsalc",
    "codelcodven",
    "enex",
    "centinela",
    "nazar",
    "astudillo",
    "spl2"
    ]
len(clientes)
```

```python
for file in files:

    title_name = ""
    for c in clientes:
        if c in file:
            title_name = c.upper()

    file_name = "stats_/" + file
    df = pd.read_csv(file_name, sep = ";")
    proveedores = df.Provider.unique()
    n_prov = len(proveedores)
    print(file, n_prov)

    if n_prov != 0:
        fact = 3.
        if n_prov == 1:
            fact = 4.

        fig, ax1 = plt.subplots(n_prov, 1, figsize=(6, fact*n_prov), sharex = True, dpi = 250)
        fig.subplots_adjust(hspace=0.225, top = 0.87)

        if n_prov == 1:
            ax1 = [ax1]

        for i, val in enumerate(proveedores):
            print("entra y hace la imagen")

            prov_df = df[df.Provider == val]

            days = prov_df.groupby(by = "Day").mean()["AvgDTM"].index[:-1]
            avgDTM = prov_df.groupby(by = "Day").mean()["AvgDTM"].values[:-1]
            avgDTS = prov_df.groupby(by = "Day").mean()["AvgDTS"].values[:-1]


            ax1[i].set_title("Proveedor: " + val, fontsize = 15)
            ax1[i].plot(days, avgDTM, "bo-", label = "DTM", alpha = 0.9)
            ax1[i].set_ylabel("Promedio DTM", fontsize = 13)
            ax1[i].set_xlabel("Días", fontsize = 13)
            ax1[i].legend(loc = "upper left")
            ax1[i].set_ylim([-0.05*avgDTM.max(),1.3*avgDTM.max()])
            ax1[i].set_xticks(days)
            ax1[i].set_xticklabels(days, rotation = 40, fontdict = {"horizontalalignment": "right"})

            ax2 = ax1[i].twinx()

            ax2.plot(days, avgDTS, "ro--", label = "DTS", alpha = 0.8)
            ax2.set_ylabel("Promedio DTS", fontsize = 13)
            ax2.set_ylim([-0.05*avgDTS.max(),1.3*avgDTS.max()])
            ax2.legend(loc = "upper right")
            ax2.set_xticklabels(days, rotation = 40, fontdict = {"horizontalalignment": "right"})

        fig.suptitle(title_name, fontsize=17, y=1.)

        plt.tight_layout()

        fig.savefig("graficos_DTX/" + title_name + ".png", bbox_inches='tight')
```
