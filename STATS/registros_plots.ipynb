```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def get_csv_file_names(directory):
    file_names = []
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)) and filename.endswith('.csv'):
            file_names.append(filename)
    return file_names
```

```python
clientes = [
    "coseducam2",
    "agrosuper_tp",
    "agrosuper4",
    "greenamerica2",
    "abastible_e",
    "abastible3",
    "sqmc3",
    "sqmpt",
    "sqmraevpp",
    "codelcodsal2",
    "codelcodsalc",
    "codelcodven",
    "enex",
    "centinela",
    "nazar",
    "astudillo",
    "spl2"
    ]
```

```python
directory = "stats_/"
files = get_csv_file_names(directory)
files
```

```python
file = directory + files[4]
df = pd.read_csv(file, sep = ";")
df.head()
```

```python
df.groupby(by = ["Day"]).sum()
```

```python
for file in files:
    file_name = directory + file
    df = pd.read_csv(file_name, sep = ";")

    title_name = ""
    for c in clientes:
        if c in file:
            title_name = c.upper()



    proveedores = df.Provider.unique()
    n_prov = len(proveedores)

    fact = 4.
    if n_prov != 1:
        fact = 3.7

    fig, ax1 = plt.subplots(n_prov, 1, sharex = True, dpi = 150, figsize=(6, 4*n_prov))
    fig.subplots_adjust(hspace=0.225, top = 0.92)

    if n_prov == 1:
        ax1 = [ax1]

    for i, val in enumerate(proveedores):

        prov_df = df[df.Provider == val]

        days = prov_df.groupby(by = "Day").sum().index[:-1]
        registros = prov_df.groupby(by = "Day").sum()["Num"].values[:-1]
        n_assets = np.array([prov_df[(prov_df["Day"] == dia) & (prov_df["Num"] != 0)]["Asset"].unique().shape[0] for dia in days])


        ax1[i].set_title("Proveedor: " + val, fontsize = 15)
        ax1[i].plot(days, registros, "bo-", label = "NÂ° Registros", alpha = 0.9)
        ax1[i].set_ylabel("# Registros", fontsize = 13)
        ax1[i].set_xlabel("DÃ­as", fontsize = 13)
        ax1[i].legend(loc = "upper left") #
        ax1[i].set_ylim([-0.05*registros.max(),1.3*registros.max()])
        ax1[i].set_xticks(days)
        ax1[i].set_xticklabels(days, rotation = 40, fontdict = {"horizontalalignment": "right"})

        ax2 = ax1[i].twinx()

        ax2.plot(days, n_assets, "ro--", label = "NÂ° Patentes", alpha = 0.8)
        ax2.set_ylabel("# Patentes", fontsize = 13)
        ax2.set_ylim([-0.05*n_assets.max(),1.3*n_assets.max()])
        ax2.legend(loc = "upper right") #
        ax2.set_xticklabels(days, rotation = 40, fontdict = {"horizontalalignment": "right"})

    fig.suptitle(title_name, fontsize=17, y=1.)

    plt.tight_layout()

    fig.savefig("graficos_registros/" + title_name + ".png", bbox_inches='tight')
```
