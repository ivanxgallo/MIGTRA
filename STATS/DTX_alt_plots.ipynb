```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def get_csv_file_names(directory):
    file_names = []
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)) and filename.endswith('.csv'):
            file_names.append(filename)
    return file_names
```

```python
clientes = [
    "coseducam2",
    "agrosuper_tp",
    "agrosuper4",
    "greenamerica2",
    "abastible_e",
    "abastible3",
    "sqmc3",
    "sqmpt",
    "sqmraevpp",
    "codelcodsal2",
    "codelcodsalc",
    "codelcodven",
    "enex",
    "centinela",
    "nazar",
    "astudillo",
    "spl2"
    ]
```

```python
directory = "stats_"
files = get_csv_file_names(directory)
files
```

```python
for file in files:

    title_name = ""
    for c in clientes:
        if c in file:
            title_name = c.upper()

    file_name = "stats_/" + file
    df = pd.read_csv(file_name, sep = ";")
    proveedores = df.Provider.unique()
    n_prov = len(proveedores)
    print(file, n_prov)

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 7), sharex = True, dpi = 250)
    fig.subplots_adjust(hspace=0.225, top = 0.87)

    DTMmax = []
    DTSmax = []


    for i, val in enumerate(proveedores):

        prov_df = df[df.Provider == val]

        days = prov_df.groupby(by = "Day").mean()["AvgDTM"].index[:-1]
        avgDTM = prov_df.groupby(by = "Day").mean()["AvgDTM"].values[:-1]
        ax1.plot(days, avgDTM, "o-", alpha = 0.9, label = val)
        DTMmax.append(avgDTM.max())

    DTMmax = np.array(DTMmax)
    ax1.set_title("DTM", fontsize = 15)
    ax1.set_ylabel("Promedio DTM", fontsize = 13)
    ax1.set_xlabel("Días", fontsize = 13)
    ax1.legend(loc = 2)
    ax1.set_ylim([-0.05*DTMmax.max(),1.3*DTMmax.max()])
    ax1.set_xticks(days)
    ax1.set_xticklabels(days, rotation = 40, fontdict = {"horizontalalignment": "right"})

    for i, val in enumerate(proveedores):

        prov_df = df[df.Provider == val]

        days = prov_df.groupby(by = "Day").mean()["AvgDTM"].index[:-1]
        avgDTS = prov_df.groupby(by = "Day").mean()["AvgDTS"].values[:-1]
        ax2.plot(days, avgDTS, "o-", alpha = 0.9, label = val)
        DTSmax.append(avgDTS.max())

    DTSmax = np.array(DTSmax)
    ax2.set_title("DTS", fontsize = 15)
    ax2.set_ylabel("Promedio DTS", fontsize = 13)
    ax2.set_ylim([-0.05*DTSmax.max(),1.3*DTSmax.max()])
    ax2.legend()
    ax2.set_xticklabels(days, rotation = 40, fontdict = {"horizontalalignment": "right"})

    fig.suptitle(title_name, fontsize=17, y=1.)

    plt.tight_layout()

    #fig.savefig("graficos_DTX_alternativa/" + title_name + ".png", bbox_inches='tight')
```
