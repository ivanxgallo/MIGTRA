```python
import json
from openpyxl import load_workbook
import pandas as pd
import numpy as np

original_filename = 'Maestro Enero 2023.xlsx'
# #Load original excel file with active sheet
o_wb = load_workbook(original_filename)
sheet = o_wb.worksheets[0]
```

```python

def analyze_changes(sheet):

    row_count = sheet.max_row
    column_count = sheet.max_column
    headers = [c.value for c in sheet[1] if c.value is not None]
    changes = []
    for i in range(1,row_count):
        #Asumiendo que name siempre este en la col 2
        cell = sheet.cell(column=2, row=i)
        bgColor = cell.fill.bgColor.index
        fgColor = cell.fill.fgColor.index
        if bgColor != '00000000' or fgColor != '00000000':
            asset_changes = []
            for idx,c in enumerate(sheet[i]):
                if idx>2:
                    _bgColor = c.fill.bgColor.index
                    _fgColor = c.fill.fgColor.index
                    if _bgColor != '00000000' or _fgColor != '00000000':
                        asset_changes.append({'value':c.value,'col':headers[c.column-1]})
            changes.append({'asset':cell.value,'changes':asset_changes})
    return changes
changes_to_apply = analyze_changes(sheet)
print(changes_to_apply[0])
current_file = 'maestro_20230110.xlsx'
c_df = pd.read_excel(current_file)
for i in changes_to_apply:
    asset = i['asset']
    match = c_df.loc[c_df['name'] == asset]
    match_row = match.index[0]
    next_values_df = match.loc[match_row:match_row+50]
    #Modificar valores actuales
    print('Actual')
    print(c_df.iloc[match_row])
    for x in i['changes']:
        c_df.loc[match_row,x['col']] = x['value'] if c_df.loc[match_row,x['col']] != x['value'] else c_df.loc[match_row,x['col']]
    print('modificado')
    print(c_df.iloc[match_row])
    # filas_sin_nombre = []
    # has_name = False
    # filas = []
    # for i, row in next_values_df.iterrows():
    #     if pd.notna(row['name']):
    #         has_name = True
    #         filas.append(row)
    #     elif has_name:
    #         filas_sin_nombre.append(row)
    #         has_name = False
    # #Siempre que consiga un match, debe modificar los cambios del arregl changes to apply sobre la fila que hizo match.
    # #Luego debe generar el historico de cambios temporales sobre las filas t_%s_ds y t_%s_de segun corresponda.
    # for x in i['changes']:
    #     c_df.loc[match_row,x['col']] = x['value']
    # if len(filas_sin_nombre)==0:
    #     c_df.index[match_row+1] = pd.


```
